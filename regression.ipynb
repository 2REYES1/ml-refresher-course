```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape) # shapes are (rows, cols)
print(X[0]) 
```

```python
X = X[:, 2] # ':' selects all rows, index 2 selects 3rd col
X = X.reshape((-1, 1)) # reshape, -1 auto conf dimension
```

```python
# 33% of data set will be train set, remaining 67% for training set
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# train X and y training sets with Linear Regression Model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# create prediction with the test data set
y_pred = model.predict(X_test)
```

```python
# create scatterplot to show data
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

The blue line shows what the model is predicting. For example, if the model predicts that a value of 0.05 BMI has a value of 200 for Disease Progression. The black scattered points are the actual values/points of the test data.
